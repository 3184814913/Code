## 4.6.4 改变图的坐标轴的取值范围

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x=np.linspace(0,10,100)

plt.plot(x,np.sin(x))
plt.xlim(11,-2)
plt.ylim(2.2,-1.3)
```

```python
plt.plot(x,np.sin(x))
plt.axis([-1,21,-1.6,1.6])
```

```python
plt.plot(x,np.sin(x))
plt.axis([-1,21,-1.6,1.6])
plt.axis("equal") 
```

## 4.6.5 去掉边界的空白

```python
plt.plot(x,np.sin(x))
plt.axis([-1,21,-1.6,1.6])
plt.axis("tight")
```

## 4.6.6 在同一个坐标上画两个图

## 4.6.9 散点图的画法

```python
from sklearn.datasets.samples_generator import make_blobs 
X,y=make_blobs(n_samples=300,centers=4,random_state=0, cluster_std=1.0)       
plt.scatter(X[:,0],X[:,1],c=y,s=50,cmap="rainbow")
```
